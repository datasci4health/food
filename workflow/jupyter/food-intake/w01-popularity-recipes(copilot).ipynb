```python
# Read a CSV file in the relative path ../../../data/food-intake/profile/commodity-profile.csv
# and print the first five rows of the file
import pandas as pd

df = pd.read_csv('../../../data/food-intake/profile/commodity-profile.csv')
print(df.head())
```

```python
# Draw a scatter plot of this file with POPULARITY vs. RECIPES,
# changing the bullet color of POPULARITY according to its value

import matplotlib.pyplot as plt

plt.scatter(df['POPULARITY'], df['RECIPES'], c=df['POPULARITY'], cmap='viridis')

plt.xlabel('POPULARITY')
plt.ylabel('RECIPES')
plt.title('POPULARITY vs. RECIPES')
plt.show()
```
