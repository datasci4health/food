```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
df = pd.read_csv('../../../data/food-intake/analysis/profile/commodity-profile.csv')

# Show the file content
df.head()
```

```python
# Create a scatter plot
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df['POPULARITY'], df['RECIPES'], c=df['POPULARITY'], cmap='viridis', s=50)


# Add labels and title
plt.xlabel('Popularity')
plt.ylabel('Recipes')
plt.title('Popularity vs Recipes Scatter Plot')

# Add a color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Popularity', fontsize=12)

# Improve the layout
plt.tight_layout()

# Show the plot
plt.show()
```
