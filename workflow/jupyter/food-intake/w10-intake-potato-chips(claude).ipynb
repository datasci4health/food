How can a write a Python + Pandas program over Jupyter to do the following:
1. read a CVS file named intake-person-demo(potato-chips).csv
2. draw a scatter plot of the columns: age vs. intake_bw, with the color of the bullets according to the value of intake_bw

*Result:*

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Read the CSV file
df = pd.read_csv('../../../data/food-intake/grouped/intake-person-demo(potato-chips).csv')

# 2. Create a scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['age'], df['intake_bw'], c=df['intake_bw'], cmap='viridis')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Intake (body weight)')
plt.title('Age vs. Intake (body weight)')

# Add a color bar
plt.colorbar(scatter, label='Intake (body weight)')

# Show the plot
plt.show()
```
